# Covariance
#### --->Covariance tells u whether two variables move together.
## Positive covariance:
#### --->If X variable increases and Y variable also increases.
## Negative covrariance:
#### --->If X variable increases and y variable decreases.
## Zero covariance:
#### --->If they move randomly
#### Note: Covariance only shows direction not strength.

# Correlation
### --->Direction + How strong the relationship.
### --->It is basically a scaled version of covariance.
## Range:
### --->Correlation always lies in between -1 to +1.
### --->+1  Perfect positive relationship(Both are same direction)
### --->0  No relationship
### --->-1  Perfect negative relationship(Both are opposite direction)
